```python
def cleanup(tweet):
    if isinstance(tweet, str):
        tw = tweet.split(" ")
        tw_edit = []
        for word in tw:
            if not "@" in word and not "RT" in word:
                word = word.lower()
                tw_edit.append(word)
        tweet = " ".join(tw_edit)
        del tw
        del tw_edit
        return tweet
    else:
        return None
```

```python
import spacy
from spacy.lang.es.examples import sentences 
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load("es_core_news_sm")
nlp.add_pipe('spacytextblob')
```

```python
def sentiment_nlp(tr, ret=1):
    # tr =tr.iloc[0]
    if isinstance(tr, str):
        # print(tr)
        aspects = [] 
        targets = []
        descriptions = []
        doc = nlp(tr)
        descriptive_term = ''
        target = ''
        for token in doc:
            if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
                target = token.text
            if token.pos_ == 'ADJ':
                prepend = ''
                for child in token.children:
                    if child.pos_ != 'ADV':
                        continue
                    prepend += child.text + ' '
                descriptive_term = prepend + token.text
            if not len(target)==0 and not (target in targets and descriptive_term in descriptions):
                targets.append(target)
                descriptions.append(descriptive_term)
                aspects.append({'aspect': target, 'description': descriptive_term, 'polarity': doc._.polarity, 'subjectivity': doc._.subjectivity, 'assesments': doc._.assessments})
        t = aspects
        del targets
        del descriptions
        del doc
        if len(t)!=0:
            # print(t)
            placeholder = {}
            placeholder['polarity'] = t[0]['polarity']
            placeholder['subjectivity'] = t[0]['subjectivity']
            placeholder['assesments'] = t[0]['assesments']
            keywords = []
            for elem in t:
                keywords.append((elem['aspect'], elem['description']))
            placeholder['contents'] = keywords
            t = placeholder
            pol = t['polarity']
            pola = ''
            if pol < 0:
                pola = 'ENOJO'
            elif pol < 0.018 and pol >= 0:
                pola = "FRUSTRACI√ìN"
            elif pol < 0.05 and pol >= 0.018:
                pola = "NEUTRAL"
            elif pol < 0.14 and pol >= 0.05:
                pola = "CONFORMIDAD"
            elif pol >= 0.14:
                pola = "SATISFACCI√ìN"
            else:
                pola = "SIN DATOS POLARIDAD"
            t['polarity ranking'] = pola

            pol = t['subjectivity']
            pola = ''
            if pol < 0.188:
                pola = 'OBJETIVO'
            elif pol < 0.41 and pol >= 0.188:
                pola = "NEUTRAL"
            elif pol >= 0.41:
                pola = "SUBJETIVO"
            else:
                pola = "SIN DATOS SUBJETIVIDAD"
            t['subjectivity ranking'] = pola
            if ret==1:
                return t['polarity ranking']
            else:
                return t['subjectivity ranking']
        else:
            return 'INDEFINIDO'
    return 'INDEFINIDO'
```

```python
df['Emoci√≥n'] = df['tweet_text'].apply(lambda tr: sentiment_nlp(tr))
```
