```python
from sentiment_analysis_spanish import sentiment_analysis
import os
from tqdm import tqdm
```

```python
def clean_transcript(text):
    prohibited = ["ok", "titular", "rut", "cédula"]
    if ":" in text:
        dummy = []
        ls = text.split('\n')
        for item in ls:
            lss = item.split(":")
            for it in lss:
                flag = True
                for w in prohibited:
                    if w in it or it in w:
                        flag = False
                if flag:
                    for c in it:
                        if not c.isdigit() and not len(c) == 0:
                            dummy.append(c)
        if not len(dummy)==0:
            return "".join(dummy)
        else:
            return text
    else:
        return text
```

```python
transcripciones = []

for root, dirs, files in os.walk('/Users/lakat/Downloads/PROB MOT SIN INTERNET'):
    for file in tqdm(files):
        if "txt" in file:
            transcript = open(file, "r", encoding='utf-8')
            content = transcript.read()
            transcript.close()
            content = clean_transcript(content)
            transcripciones.append(content)
            
```

```python
import spacy
from spacy.lang.es.examples import sentences 
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load("es_core_news_sm")
nlp.add_pipe('spacytextblob')

trans_aspects = []

for tr in transcripciones:
    aspects = [] 
    targets = []
    descriptions = []
    doc = nlp(tr)
    descriptive_term = ''
    target = ''
    for token in doc:
        if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
            target = token.text
        if token.pos_ == 'ADJ':
            prepend = ''
            for child in token.children:
                if child.pos_ != 'ADV':
                    continue
                prepend += child.text + ' '
            descriptive_term = prepend + token.text
        if not len(target)==0 and not (target in targets and descriptive_term in descriptions):
            targets.append(target)
            descriptions.append(descriptive_term)
            aspects.append({'aspect': target, 'description': descriptive_term, 'polarity': doc._.polarity, 'subjectivity': doc._.subjectivity, 'assesments': doc._.assessments})
    trans_aspects.append(aspects)
```

```python
new_transcriptions = []
for t in trans_aspects:
    placeholder = {}
    placeholder['polarity'] = t[0]['polarity']
    placeholder['subjectivity'] = t[0]['subjectivity']
    placeholder['assesments'] = t[0]['assesments']
    keywords = []
    for elem in t:
        keywords.append((elem['aspect'], elem['description']))
    placeholder['contents'] = keywords
    new_transcriptions.append(placeholder)
```
